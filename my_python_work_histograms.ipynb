```python
import pandas as pd
file_name= 'E:/Python20230505/Acid_mydata.csv'
data=pd.read_csv(file_name)
data
```

```python
Acid_2022=data['Acid_2022']
Acid_2021=data['Acid_2021']
```

```python
col_array1=Acid_2022.to_numpy()
col_array1
```

```python
col_array2=Acid_2021.to_numpy()
col_array2
```

```python
plt.figure(figsize=(8,6)) #图片大小
plt.hist(col_array1,bins=22,label='Acid_2022',histtype='step',density=True,lw=2,rwidth=1) #hist 直方图 label #rwidth用于设置每个bin中的bar相对bin的大小
plt.hist(col_array2,bins=22,label='Acid_2021',histtype='step',density=True ,linewidth=2,rwidth=1)#histtype 直方图类型，density=True 归一化
plt.xlabel('Acid',fontsize=18) #fontsize=18字号 #lw=linewidth
plt.ylabel('Density',fontsize=18)
plt.tick_params(axis='both',labelsize=16)#可以不是‘both’
plt.legend(loc='upper right',fontsize=15,ncol=1)#loc 位置 ncol几列
plt.xlim((0, 1.2))# xlim 设置x坐标轴的范围 xticks设置间距，这块不是很明白需要再研究一下，还有一个import MultipleLocator设置间距的注意参考
plt.xticks=np.arange(0,0.8,0.1)
plt.show()
```

```python
textstr1 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(col_array1)),
    r'$\sigma_b=%.4f$' % (np.std(col_array2))))
#标准差-
```

```python
fig, axes=plt.subplots(1,3,figsize=(20,4))#注意此处作图和上面的有所不同 将图看成矩阵1行3列   figsize是整体的大小调整
ax= axes[0] #将下面的三个图看做0,1,2
ax.hist(col_array1,bins=22,label='Acid_2022',histtype='step',density=True,lw=2,rwidth=1)
ax.hist(col_array2,bins=22,label='Acid_2021',histtype='step',density=True ,lw=2,rwidth=1)
ax.set_xlabel('Acid',fontsize=18)
ax.set_ylabel('Density',fontsize=18)
ax.tick_params(axis='both',labelsize=16)
ax.legend(loc='upper right',fontsize=12,ncol=1)
ax.set_xlim((0, 1.4))



ax=axes[1]
ax.hist(col_array1,bins=23,label='Acid_2022',rwidth=1)
ax.set_xlim((0.02, 1.2))
ax.legend(loc='upper right',fontsize=12,ncol=1)
ax.set_ylabel('Count',fontsize=18)
ax.set_xlabel('Acid',fontsize=18)

ax=axes[2]
ax.hist(col_array2,bins=22,label='Acid_2021',facecolor='green',rwidth=1)
ax.set_xlim((0.02, 1.4))
ax.set_ylim((0,18))
ax.legend(loc='upper right',fontsize=12,ncol=1)
ax.set_ylabel('Count',fontsize=18)
ax.set_xlabel('Acid',fontsize=18)

plt.savefig("Acid_mydata_20230505.pdf")
```

```python
from matplotlib.pyplot import MultipleLocator#从pyplot导入MultipleLocator类，这个类用于设置刻度间隔
x_major_locator=MultipleLocator(0.1) #把x轴的刻度间隔设置为0.1，并存在变量里
ax.xaxis.set_major_locator(x_major_locator)#把x轴的主刻度设置为0.1的倍数
```
