```python
import pandas as pd
file_name= 'E:/Python20230505/MD-histogram.csv'
data=pd.read_csv(file_name)
data
```

```python
data_MD_2022=data['MD2022']
MD_2022=data_MD_2022.dropna(how='any')
MD_2022
arr2=MD_2022.to_numpy()
arr2
```

```python
import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Times New Roman'
import seaborn as sns
sns.set_theme(style="darkgrid")
# 去除极值 
arr2 = arr2[3:-3]

# Box-Cox变换 
z, lmbda = stats.boxcox(arr2)
z0 = stats.boxcox(arr2, lmbda=0)   # log变换
z05 = stats.boxcox(arr2, lmbda=0.5)

# 选择最佳变换参数
skew_z, kurt_z = stats.skew(z), stats.kurtosis(z)  
skew_z0, kurt_z0 = stats.skew(z0), stats.kurtosis(z0)
skew_z05, kurt_z05 = stats.skew(z05), stats.kurtosis(z05)
if (np.abs(skew_z) < 2) & (kurt_z < 7): 
    lz = z 
elif (np.abs(skew_z0) < 2) & (kurt_z0 < 7):  
    lz = z0 
elif (np.abs(skew_z05) < 2) & (kurt_z05 < 7):  
    lz = z05

# 标准化
z_new = (lz - lz.mean())/lz.std()  

# 作图  
sns.set_theme(style="darkgrid")
plt.rcParams['font.family'] = 'Times New Roman'
fig, ax = plt.subplots(2,2)
stats.probplot(arr2, dist="norm", plot=ax[0,0])  
ax[0,0].set_title('Statistic_data')
stats.probplot(lz, dist="norm", plot=ax[0,1])  
ax[0,1].set_title('Box-Cox')
ax[1,0].hist(arr2,bins=15, rwidth=1, color='blue', alpha=0.5)  
ax[1,1].hist(z_new, bins=15,rwidth=1, color='green', alpha=0.5)
plt.subplots_adjust(wspace=0.4, hspace=0.4) 
plt.show()
```
